```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('Books-BX-10.csv')
books_length = len(books)

ISBNs = books['ISBN'].tolist()
```

```python
# Mapping ISBN representation
book_to_vector = {}

for i in range(books_length):
  book_id = ISBNs[i]
  book_vector = np.zeros(books_length)
  book_vector[i] = 1
  book_to_vector[book_id] = book_vector
```

```python
# Create user representation by summing the respective vectors
train_set_8 = pd.read_csv('train_set_8.csv')
# Mapping user id to a list of ISBNs
user_to_isbn_list = train_set_8.groupby('User-ID')['ISBN'].apply(list).to_dict()

user_representations = {}
for key in user_to_isbn_list.keys():
  book_id_list = user_to_isbn_list[key]
  user = np.zeros(books_length)
  # Sum all the book vectors to create user representation
  for user_book in book_id_list:
    user+=book_to_vector[user_book]
  # Add to the user representations map
  user_representations[key] = user

# Save the file
user_representation = pd.DataFrame.from_dict(user_representations, orient='index')
user_representation.to_csv('train_8_user_representation_agg.csv', index_label='User-ID')
```

```python
# Same Process for the test set users

# Create user representation by summing the respective vectors
test_set_8 = pd.read_csv('test_set_8.csv')
# Mapping user id to a list of ISBNs
user_to_isbn_list_test = test_set_8.groupby('User-ID')['ISBN'].apply(list).to_dict()

user_representations_test = {}
for key in user_to_isbn_list_test.keys():
  book_id_list = user_to_isbn_list_test[key]
  user = np.zeros(books_length)
  # Sum all the book vectors to create user representation
  for user_book in book_id_list:
    user+=book_to_vector[user_book]
  # Add to the user representations map
  user_representations_test[key] = user

# Save the file
user_representation_test = pd.DataFrame.from_dict(user_representations_test, orient='index')
user_representation_test.to_csv('test_8_user_representation_agg.csv', index_label='User-ID')
```

```python
# Now we have the user representation based on the one-hot encoding of the 8 books they've read
# We want to add the age feature to it - this will be the full user vector we pass to the encoder we want to train
user_table = pd.read_csv('Users_BX_10_5.csv')

full_user_train = user_representation.reset_index(names='User-ID').merge(user_table,on='User-ID',how='inner')
full_user_test = user_representation_test.reset_index(names='User-ID').merge(user_table,on='User-ID',how='inner')
```

```python
# Save to csv
full_user_test.to_csv('full_user_test_rep.csv',index = False)
full_user_train.to_csv('full_user_train_rep.csv',index = False)
```
