```python
import pandas as pd

file_path = "Books_with_Visual_Features.csv"
df = pd.read_csv(file_path, delimiter=",", encoding="latin1")
```

```python
print(df.head())
```

```python
from sentence_transformers import SentenceTransformer
from tqdm import tqdm

model = SentenceTransformer('all-MiniLM-L6-v2')

tqdm.pandas()
df['Title_Embedding'] = df['Title'].progress_apply(lambda title: model.encode(title))

title_embedding_dim = len(df['Title_Embedding'].iloc[0])
title_embedding_cols = [f"Title_Embed_{i}" for i in range(title_embedding_dim)]

df_emb = pd.DataFrame(df['Title_Embedding'].to_list(), columns=title_embedding_cols, index=df.index)

df = pd.concat([df, df_emb], axis=1)

df.drop(columns=['Title_Embedding'], inplace=True)
```

```python
print(df.head())
```

```python
output_path = "Books_BX_10_5_with_text-embeddings.csv"
df.to_csv(output_path, index=False)
```
