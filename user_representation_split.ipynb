```python
import numpy as np
import pandas as pd
```

```python
test_set = pd.read_csv('test_set.csv')
train_set  = pd.read_csv('train_set.csv')
```

```python
# Select top 10 ratings for each user
test_top10 = test_set.groupby('User-ID').filter(lambda x: len(x)>=10).groupby('User-ID',group_keys=False).apply(lambda x : x.nlargest(10,'Rating'))
train_top10  = train_set.groupby('User-ID').filter(lambda x: len(x)>=10).groupby('User-ID',group_keys=False).apply(lambda x : x.nlargest(10,'Rating'))
display(test_top10.head(30))
```

```python
# 10 check
counts = train_top10.groupby('User-ID').size().reset_index(name='count')
display(len(counts[counts['count']<10]))
counts = test_top10.groupby('User-ID').size().reset_index(name='count')
display(len(counts[counts['count']<10]))
```

```python
# Randomly sample 2 books per user - for both 
test_set_2 = test_top10.groupby('User-ID',group_keys=False).apply( lambda x: x.sample(2,random_state=42))
# display(test_set_2.head(6))
test_set_8 = test_top10.drop(test_set_2.index)
# display(test_set_8.head(24))

train_set_2 = train_top10.groupby('User-ID',group_keys=False).apply( lambda x: x.sample(2,random_state=42))
train_set_8 = train_top10.drop(train_set_2.index)
```

```python
# Save the files

train_set_2.to_csv('train_set_2.csv',index=False)
train_set_8.to_csv('train_set_.csv',index= False)

test_set_2.to_csv('test_set_2.csv',index = False)
test_set_8.to_csv('test_set_8.csv',index = False)
```
