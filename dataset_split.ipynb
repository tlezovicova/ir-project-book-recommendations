```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv(f"Ratings-BX-10.csv")
```

```python
counts = ratings['User-ID'].value_counts()
print(counts)
```

```python
# Get 20% of ratings but keep users grouped by id to avoid overflow
test_size = int(0.2* len(ratings))
shuffled_ids = np.random.permutation(counts.index)
test_user_ids = []
num_records = 0
for user in shuffled_ids:
    test_user_ids.append(user)
    num_records+=counts[user]
    if num_records >= test_size:
        break
```

```python
mask = ratings['User-ID'].isin(test_user_ids)
test_set_ratings = ratings[mask]
train_set_ratings = ratings[~mask]
print(len(test_set_ratings),len(train_set_ratings))
```

```python
books = pd.read_csv("Books_BX_10_5_with_text-embeddings.csv")
users = pd.read_csv("Users_BX_10_5.csv")
```

```python
def bin_ratings(rating):
  if rating < 3:
    return 0
  elif rating < 5:
    return 1
  elif rating < 7:
    return 2
  elif rating < 9:
    return 3
  elif rating <=10:
    return 4
  else:
    return -1
```

```python
# Test set enrichment
t1_ratings = test_set_ratings.merge(books,how='inner',on='ISBN')
final_test = t1_ratings.merge(users,how='inner',on='User-ID')
# Bin Rating to relevance
final_test['Relevance'] = final_test['Rating'].apply(bin_ratings)
display(final_test.head(20))
final_test.to_csv("test_set.csv",index=False)
```

```python
# Train set enrichment
t2_ratings = train_set_ratings.merge(books,how='inner',on='ISBN')
final_train = t2_ratings.merge(users,how='inner',on='User-ID')
# Bin Rating to relevance
final_train['Relevance'] = final_train['Rating'].apply(bin_ratings)
display(final_train.head(20))
final_train.to_csv("train_set.csv",index=False)
```
